```python
# Poner todo junto - Hacer un loop for a través de los resultados y añadir datos dentro de una lista
modelo = []
kilometraje = []
año = []
nom_distribuidor = []
precio = []

for result in result_items:
    
    # modelo
    modelo.append(result['model'])
    
    # kilometraje
    kilometraje.append(result['mileage'])
    
    # año
    año.append(result['year']) 
    
    # nombre de distribuidor
    nom_distribuidor.append(result['dealer_name']) 
    
    # precio
    precio.append(result['price'])
```

```python
# crear Dataframe - una pagina
df = pd.DataFrame({'Modelo': modelo, 'Kilometraje':kilometraje, 'Año':año, 
                   'Nom_distribuidor': nom_distribuidor,'Precio': precio})
```
